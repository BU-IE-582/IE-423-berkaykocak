We suppose that pixel values have normal distribution. Because, histogram is bell-shaped; Q-Q plot, empirical and theoretical CDFs and P-P plot are well fitted.

```r
mysample<-sample(length(imggray[,,1]),512,replace=TRUE)
shapiro.test(imggray[,,1][mysample])
```

To double-check, we applied Shapiro-Wilk Normality Test and obtained a p-value higher than 0.05. Therefore, we can say that pixel values are normally distributed.

### 2)

```r
mean(imggray[,,1])
sd(imggray[,,1])
```

Firstly, we estimated the mean by calculating sample mean and the variance by calculating sample variance. It is known that sample mean and sample variance are unbiased estimators of mean and variance.

```r
mean_mle<-function(array)
{
        mean(array)
}

sd_mle<-function(array)
{
        sqrt(sum((array-mean_mle(array))^2)/(length(array)))
}

mean_mle(imggray[,,1])
sd_mle(imggray[,,1])
```

MLE is another way to estimate those parameters. We created two functions to calculate mean and standard deviation via MLE approach. Here are the formulas:

$$\hat{\mu}_{MLE} = \dfrac{1}{n}\sum\limits_{i=1}^n x_i $$
$$ $$
$$\hat{\sigma}_{MLE} = \sqrt{\dfrac{1}{n}\sum\limits_{i=1}^n (x_i-\mu)^2}$$

As it can be seen above, mean estimation of MLE approach and sample mean are equal. Besides, standart deviation estimates are quite close. For the sake of simplicity, we decided to use sample variance as the estimator of standart deviation.
